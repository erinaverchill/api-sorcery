```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
epi = pd.read_csv('../Resources/economic_personal_income_clean.csv')
rec = pd.read_csv('../Resources/regional_economic_analysis_clean.csv')
```

```python
# Industries that are most popular amongst each education level 
# median income in each of these industries (according to education)
rec.head()
```

```python
hs = rec[rec['Education Level'] == 'High school diploma or equivalent']
hs.head()
```

```python
hs = hs.drop(['Area Type', 'Area Name', 'SOC Code', 'Median Hourly Wage', 'New Job Openings', 'Replacement Job Openings', 'Total Projected Job openings.1'], axis=1)
```

```python
hs
```

```python
hs_top_industry = hs["Industry Cluster"].value_counts()
hs_top_industry = pd.DataFrame(hs_top_industry)
hs_top_industry = hs_top_industry.head()
```

```python
hs_industry_matplotlib = hs_top_industry.plot.bar()
plt.xticks(rotation=45)
```

```python
hs_top_jobs = hs['SOC Title'].value_counts()
hs_top_jobs = pd.DataFrame(hs_top_jobs)
hs_top_jobs = hs_top_jobs.head()
hs_top_jobs
```

```python
hs_jobs_matplotlib = hs_top_jobs.plot.bar()
```

```python
office_clerks = hs[hs['SOC Title'] == 'Office Clerks, General']
office_clerks = office_clerks["Median Annual Wage"].median(axis=0)
office_clerks = pd.DataFrame({"Wage (Mode)":[office_clerks]})
office_clerks = office_clerks.rename(index={0: "Office Clerks"})
office_clerks
```

```python
Sec_AA = hs[hs['SOC Title'] == 'Secretaries and Administrative Assistants, Except Legal, Medical, and Executive']
Sec_AA = Sec_AA["Median Annual Wage"].mode()
Sec_AA = pd.DataFrame(Sec_AA, columns=["Wage (Mode)"])
Sec_AA = Sec_AA.rename(index={0: "Secretaries and Administrative Assistants"})
Sec_AA
```

```python
bookkeeping = hs[hs['SOC Title'] == 'Bookkeeping, Accounting, and Auditing Clerks']
bookkeeping = bookkeeping["Median Annual Wage"].mode()
bookkeeping = pd.DataFrame(bookkeeping, columns=["Wage (Mode)"])
bookkeeping = bookkeeping.rename(index={0: "Bookkeeping, Accounting, and Auditing Clerks"})
bookkeeping
```

```python
csr = hs[hs['SOC Title'] == 'Customer Service Representatives']
csr = csr["Median Annual Wage"].mode()
csr = pd.DataFrame(csr, columns=["Wage (Mode)"])
csr = csr.rename(index={0: "Customer Service Representatives"})
csr
```

```python
supervisor = hs[hs['SOC Title'] == 'First-Line Supervisors of Office and Administrative Support Workers']
supervisor = supervisor["Median Annual Wage"].mode()
supervisor = pd.DataFrame(csr, columns=["Wage (Mode)"])
supervisor = supervisor.rename(index={0: "Supervisors of Office and Admin. Support Workers"})
supervisor
```

```python
frames = [office_clerks, Sec_AA, bookkeeping, csr, supervisor]
result = pd.concat(frames)
result
```

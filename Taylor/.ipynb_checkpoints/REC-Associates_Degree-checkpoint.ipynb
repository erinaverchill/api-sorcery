```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rec = pd.read_csv('../Resources/regional_economic_analysis.csv')
```

```python
# Industries that are most popular amongst each education level 
# median income in each of these industries (according to education)
rec.head()
```

```python
# Re-ordering rec dataframe columns
rec = rec[["Entry Level Education","Area Type","Area Name",
           "Industry Cluster","SOC Code","SOC Title","Median Hourly Wage","Median Annual Wage","Period",
          "New Job Openings","Replacement Job Openings","Total Projected Job openings","Total Projected Job openings",
          "Online Job Ads (HWOL)"]]
rec.head()
```

```python
# All education levels in rec dataset
column_values = rec["Entry Level Education"].values.ravel()
unique_values =  pd.unique(column_values)
unique_values
```

```python
ad = rec[rec['Entry Level Education'] == "Associate's degree"]
ad.head()
```

```python
ad = ad.drop(['Area Type', 'Area Name', 'SOC Code', 'Median Hourly Wage', 
                'New Job Openings', 'Replacement Job Openings', 'Total Projected Job openings'], axis=1)
```

```python
ad
```

```python
ad_top_industry = ad["Industry Cluster"].value_counts()
ad_top_industry = pd.DataFrame(ad_top_industry)
ad_top_industry = ad_top_industry.head()
```

```python
ad_industry_matplotlib = ad_top_industry.plot.bar()
plt.xticks(rotation=45)
```

```python
ad_top_jobs = ad['SOC Title'].value_counts()
ad_top_jobs = pd.DataFrame(ad_top_jobs)
ad_top_jobs = ad_top_jobs.head()
ad_top_jobs
```

```python
ad_jobs_matplotlib = ad_top_jobs.plot.bar()
```

```python
web_developers = ad[ad['SOC Title'] == 'Web Developers']
web_developers = web_developers["Median Annual Wage"].median(axis=0)
web_developers = pd.DataFrame({"Wage (Median)":[web_developers]})
web_developers = web_developers.rename(index={0: "Web Developers"})
web_developers
```

```python
nss = ad[ad['SOC Title'] == 'Computer Network Support Specialists']
nss = nss["Median Annual Wage"].median(axis=0)
nss = pd.DataFrame({"Wage (Median)":[nss]})
nss = nss.rename(index={0: "Computer Network Support Specialists"})
nss
```

```python
eet = ad[ad['SOC Title'] == 'Electrical and Electronics Engineering Technicians']
eet = eet["Median Annual Wage"].median(axis=0)
eet = pd.DataFrame({"Wage (Median)":[eet]})
eet = eet.rename(index={0: "Electrical and Electronics Engineering Technicians"})
eet
```

```python
rn = ad[ad['SOC Title'] == 'Registered Nurses']
rn = rn["Median Annual Wage"].median(axis=0)
rn = pd.DataFrame({"Wage (Median)":[rn]})
rn = rn.rename(index={0: "Registered Nurses"})
rn
```

```python
pre_k = ad[ad['SOC Title'] == 'Preschool Teachers, Except Special Education']
pre_k = pre_k["Median Annual Wage"].median(axis=0)
pre_k = pd.DataFrame({"Wage (Median)":[pre_k]})
pre_k = pre_k.rename(index={0: "Preschool Teachers, Except Special Education"})
pre_k
```

```python
frames = [web_developers, nss, eet, rn, pre_k]
result = pd.concat(frames)
result_format = {'Wage (Median)':'${0:,.2f}'}
result.style.format(result_format)
```

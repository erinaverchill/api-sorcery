```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO
from matplotlib.legend import Legend
from matplotlib import rc
```

```python
rec = pd.read_csv('../Resources/regional_economic_analysis.csv')
```

```python
rec = rec[["Entry Level Education","Area Type","Area Name",
           "Industry Cluster","SOC Code","SOC Title","Median Hourly Wage","Median Annual Wage","Period",
          "New Job Openings","Replacement Job Openings","Total Projected Job openings","Total Projected Job openings",
          "Online Job Ads (HWOL)"]]
rec.head()
```

```python
for column in rec.columns:
    print(column)
```

```python
rec = rec.loc[:,~rec.columns.duplicated()]
rec.head()
```

```python
edu_rec = rec.groupby('Entry Level Education')["Total Projected Job openings"].sum()
edu_rec = pd.DataFrame(edu_rec)
edu_rec
```

```python
wage_rec = rec.groupby('Entry Level Education')["Median Annual Wage"].median()
wage_rec = pd.DataFrame(wage_rec)
wage_rec
```

```python
money_rec = pd.merge(wage_rec,edu_rec,on="Entry Level Education")
money_rec = money_rec.sort_values(by=['Median Annual Wage'],ascending=False)
money_format = {'Total Projected Job openings':'{0:,.0f}','Median Annual Wage':'${0:,.2f}'}
money_rec.style.format(money_format)
```

```python
fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot() # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

lns1 = money_rec['Total Projected Job openings'].plot(kind='bar', color='red', ax=ax, width=.2, position=1)
lns2 = money_rec['Median Annual Wage'].plot(kind='bar', color='blue', ax=ax2, width=.2, position=0)

fig.legend([lns1,lns2],loc='upper center')

ax.set_ylabel('Projected Job Openings')
ax2.set_ylabel('Median Annual Wage')
plt.xticks(rotation=45)
plt.show()
```

```python
m = money_rec.plot( kind= 'bar' , secondary_y= 'Total Projected Job openings' )

plt.xticks(rotation=24)
plt.show()
```

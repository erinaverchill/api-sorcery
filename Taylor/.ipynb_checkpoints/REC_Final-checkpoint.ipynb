```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rec = pd.read_csv('../Resources/regional_economic_analysis.csv')
```

```python
rec = rec[["Entry Level Education","Area Type","Area Name",
           "Industry Cluster","SOC Code","SOC Title","Median Hourly Wage","Median Annual Wage","Period",
          "New Job Openings","Replacement Job Openings","Total Projected Job openings","Total Projected Job openings",
          "Online Job Ads (HWOL)"]]
rec.head()
```

```python
for column in rec.columns:
    print(column)
```

```python
rec = rec.loc[:,~rec.columns.duplicated()]
rec.head()
```

```python
edu_rec = rec.groupby('Entry Level Education')["Total Projected Job openings"].sum()
edu_rec = pd.DataFrame(edu_rec)
edu_rec
```

```python
wage_rec = rec.groupby('Entry Level Education')["Median Annual Wage"].median()
wage_rec = pd.DataFrame(wage_rec)
wage_rec
```

```python
money_rec = pd.merge(wage_rec,edu_rec,on="Entry Level Education")
money_rec = money_rec.sort_values(by=['Median Annual Wage'],ascending=False)
money_format = {'Total Projected Job openings':'{0:,.0f}','Median Annual Wage':'${0:,.2f}'}
money_rec.style.format(money_format)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rec = pd.read_csv('../Resources/regional_economic_analysis.csv')
```

```python
# Industries that are most popular amongst each education level 
# median income in each of these industries (according to education)
rec.head()
```

```python
# Re-ordering rec dataframe columns
rec = rec[["Entry Level Education","Area Type","Area Name",
           "Industry Cluster","SOC Code","SOC Title","Median Hourly Wage","Median Annual Wage","Period",
          "New Job Openings","Replacement Job Openings","Total Projected Job openings","Total Projected Job openings",
          "Online Job Ads (HWOL)"]]
rec.head()
```

```python
# All education levels in rec dataset
column_values = rec["Entry Level Education"].values.ravel()
unique_values =  pd.unique(column_values)
unique_values
```

```python
lhs = rec[rec['Entry Level Education'] == 'Less than high school']
lhs.head()
```

```python
lhs = lhs.drop(['Area Type', 'Area Name', 'SOC Code', 'Median Hourly Wage', 
                'New Job Openings', 'Replacement Job Openings', 'Total Projected Job openings'], axis=1)
```

```python
lhs
```

```python
lhs_top_industry = lhs["Industry Cluster"].value_counts()
lhs_top_industry = pd.DataFrame(lhs_top_industry)
lhs_top_industry = lhs_top_industry.head()
```

```python
lhs_industry_matplotlib = lhs_top_industry.plot.bar()
plt.xticks(rotation=45)
```

```python
lhs_top_jobs = lhs['SOC Title'].value_counts()
lhs_top_jobs = pd.DataFrame(lhs_top_jobs)
lhs_top_jobs = lhs_top_jobs.head()
lhs_top_jobs
```

```python
lhs_jobs_matplotlib = lhs_top_jobs.plot.bar()
```

```python
laborers_and_freight = lhs[lhs['SOC Title'] == 'Laborers and Freight, Stock, and Material Movers, Hand']
laborers_and_freight = laborers_and_freight["Median Annual Wage"].median(axis=0)
laborers_and_freight = pd.DataFrame({"Wage (Median)":[laborers_and_freight]})
laborers_and_freight = laborers_and_freight.rename(index={0: "laborers_and_freight"})
laborers_and_freight
```

```python
janitors_and_cleaners = lhs[lhs['SOC Title'] == 'Janitors and Cleaners, Except Maids and Housekeeping Cleaners']
janitors_and_cleaners = janitors_and_cleaners["Median Annual Wage"].median(axis=0)
janitors_and_cleaners = pd.DataFrame({"Wage (Median)":[janitors_and_cleaners]})
janitors_and_cleaners = janitors_and_cleaners.rename(index={0: "Janitors and Cleaners"})
janitors_and_cleaners
```

```python
clerks_fillers = lhs[lhs['SOC Title'] == 'Stock Clerks and Order Fillers']
clerks_fillers = clerks_fillers["Median Annual Wage"].median(axis=0)
clerks_fillers = pd.DataFrame({"Wage (Median)":[clerks_fillers]})
clerks_fillers = clerks_fillers.rename(index={0: "Stock Clerks and Order Fillers"})
clerks_fillers
```

```python
salespersons = lhs[lhs['SOC Title'] == 'Retail Salespersons']
salespersons = salespersons["Median Annual Wage"].median(axis=0)
salespersons = pd.DataFrame({"Wage (Median)":[salespersons]})
salespersons = salespersons.rename(index={0: "Retail Salespersons"})
salespersons
```

```python
Cashiers = lhs[lhs['SOC Title'] == 'Cashiers']
Cashiers = Cashiers["Median Annual Wage"].median(axis=0)
Cashiers = pd.DataFrame({"Wage (Median)":[Cashiers]})
Cashiers = Cashiers.rename(index={0: "Cashiers"})
Cashiers
```

```python
frames = [laborers_and_freight, janitors_and_cleaners, clerks_fillers, salespersons, Cashiers]
result = pd.concat(frames)
result
```

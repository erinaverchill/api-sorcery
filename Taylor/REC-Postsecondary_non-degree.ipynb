```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rec = pd.read_csv('../Resources/regional_economic_analysis.csv')
```

```python
# Industries that are most popular amongst each education level 
# median income in each of these industries (according to education)
rec.head()
```

```python
# Re-ordering rec dataframe columns
rec = rec[["Entry Level Education","Area Type","Area Name",
           "Industry Cluster","SOC Code","SOC Title","Median Hourly Wage","Median Annual Wage","Period",
          "New Job Openings","Replacement Job Openings","Total Projected Job openings","Total Projected Job openings",
          "Online Job Ads (HWOL)"]]
rec.head()
```

```python
# All education levels in rec dataset
column_values = rec["Entry Level Education"].values.ravel()
unique_values =  pd.unique(column_values)
unique_values
```

```python
pnd = rec[rec['Entry Level Education'] == "Postsecondary non-degree award"]
pnd.head()
```

```python
pnd = pnd.drop(['Area Type', 'Area Name', 'SOC Code', 'Median Hourly Wage', 
                'New Job Openings', 'Replacement Job Openings', 'Total Projected Job openings'], axis=1)
```

```python
pnd
```

```python
pnd_top_industry = pnd["Industry Cluster"].value_counts()
pnd_top_industry = pd.DataFrame(pnd_top_industry)
pnd_top_industry = pnd_top_industry.head()
```

```python
pnd_industry_matplotlib = pnd_top_industry.plot.bar()
plt.xticks(rotation=45)
```

```python
pnd_top_jobs = pnd['SOC Title'].value_counts()
pnd_top_jobs = pd.DataFrame(pnd_top_jobs)
pnd_top_jobs = pnd_top_jobs.head()
pnd_top_jobs
```

```python
pnd_jobs_matplotlib = pnd_top_jobs.plot.bar()
```

```python
supervisor = pnd[pnd['SOC Title'] == 'First-Line Supervisors of Production and Operating Workers']
supervisor = supervisor["Median Annual Wage"].median(axis=0)
supervisor = pd.DataFrame({"Wage (Median)":[supervisor]})
supervisor = supervisor.rename(index={0: "Production & Worker Supervisors"})
supervisor
```

```python
trucks = pnd[pnd['SOC Title'] == 'Heavy and Tractor-Trailer Truck Drivers']
trucks = trucks["Median Annual Wage"].median(axis=0)
trucks = pd.DataFrame({"Wage (Median)":[trucks]})
trucks = trucks.rename(index={0: "Truck Drivers"})
trucks
```

```python
home_appliance = pnd[pnd['SOC Title'] == 'Heating, Air Conditioning, and Refrigeration Mechanics and Installers']
home_appliance = home_appliance["Median Annual Wage"].median(axis=0)
home_appliance = pd.DataFrame({"Wage (Median)":[home_appliance]})
home_appliance = home_appliance.rename(index={0: "Home Appliance Mechanics/Installers"})
home_appliance
```

```python
ecir = pnd[pnd['SOC Title'] == 'Electrical and Electronics Repairers, Commercial and Industrial Equipment']
ecir = ecir["Median Annual Wage"].median(axis=0)
ecir = pd.DataFrame({"Wage (Median)":[ecir]})
ecir = ecir.rename(index={0: "Electrical, Commercial, & Industrial Repairers"})
ecir
```

```python
aud = pnd[pnd['SOC Title'] == 'Audio and Video Equipment Technicians']
aud = aud["Median Annual Wage"].median(axis=0)
aud = pd.DataFrame({"Wage (Median)":[aud]})
aud = aud.rename(index={0: "Audio/Video Equipment Technicians"})
aud
```

```python
frames = [supervisor, trucks, home_appliance, ecir, aud]
result = pd.concat(frames)
result_format = {'Wage (Median)':'${0:,.2f}'}
result.style.format(result_format)
```

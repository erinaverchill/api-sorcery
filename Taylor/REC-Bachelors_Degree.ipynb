```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rec = pd.read_csv('../Resources/regional_economic_analysis.csv')
```

```python
# Industries that are most popular amongst each education level 
# median income in each of these industries (according to education)
rec.head()
```

```python
# Re-ordering rec dataframe columns
rec = rec[["Entry Level Education","Area Type","Area Name",
           "Industry Cluster","SOC Code","SOC Title","Median Hourly Wage","Median Annual Wage","Period",
          "New Job Openings","Replacement Job Openings","Total Projected Job openings","Total Projected Job openings",
          "Online Job Ads (HWOL)"]]
rec.head()
```

```python
# All education levels in rec dataset
column_values = rec["Entry Level Education"].values.ravel()
unique_values =  pd.unique(column_values)
unique_values
```

```python
bd = rec[rec['Entry Level Education'] == "Bachelor's degree"]
bd.head()
```

```python
bd = bd.drop(['Area Type', 'Area Name', 'SOC Code', 'Median Hourly Wage', 
                'New Job Openings', 'Replacement Job Openings', 'Total Projected Job openings'], axis=1)
```

```python
bd
```

```python
bd_top_industry = bd["Industry Cluster"].value_counts()
bd_top_industry = pd.DataFrame(bd_top_industry)
bd_top_industry = bd_top_industry.head()
```

```python
bd_industry_matplotlib = bd_top_industry.plot.bar()
plt.xticks(rotation=45)
```

```python
bd_top_jobs = bd['SOC Title'].value_counts()
bd_top_jobs = pd.DataFrame(bd_top_jobs)
bd_top_jobs = bd_top_jobs.head()
bd_top_jobs
```

```python
bd_jobs_matplotlib = bd_top_jobs.plot.bar()
```

```python
managers = bd[bd['SOC Title'] == 'General and Operations Managers']
managers = managers["Median Annual Wage"].median(axis=0)
managers = pd.DataFrame({"Wage (Median)":[managers]})
managers = managers.rename(index={0: "General and Operations Managers"})
managers
```

```python
acct = bd[bd['SOC Title'] == 'Accountants and Auditors']
acct = acct["Median Annual Wage"].median(axis=0)
acct = pd.DataFrame({"Wage (Median)":[acct]})
acct = acct.rename(index={0: "Accountants and Auditors"})
acct
```

```python
fin_man = bd[bd['SOC Title'] == 'Financial Managers']
fin_man = fin_man["Median Annual Wage"].median(axis=0)
fin_man = pd.DataFrame({"Wage (Median)":[fin_man]})
fin_man = fin_man.rename(index={0: "Financial Managers"})
fin_man
```

```python
marketing = bd[bd['SOC Title'] == 'Market Research Analysts and Marketing Specialists']
marketing = marketing["Median Annual Wage"].median(axis=0)
marketing = pd.DataFrame({"Wage (Median)":[marketing]})
marketing = marketing.rename(index={0: "Market Research Analysts and Marketing Specialists"})
marketing
```

```python
sales_man = bd[bd['SOC Title'] == 'Sales Managers']
sales_man = sales_man["Median Annual Wage"].median(axis=0)
sales_man = pd.DataFrame({"Wage (Median)":[sales_man]})
sales_man = sales_man.rename(index={0: "Sales Managers"})
sales_man
```

```python
frames = [managers, acct, fin_man, marketing, sales_man]
result = pd.concat(frames)
result_format = {'Wage (Median)':'${0:,.2f}'}
result.style.format(result_format)
```

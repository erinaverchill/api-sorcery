```python
#create A bar graph based off model found on https://matplotlib.org/index.html

labels = ['No Income', '$5,000 to $9,999', '$10,000 to $14,999','$15,000 to $24,999',
          '$25,000 to $34,999', '$35,000 to $49,999', '$50,000 to $74,999', '$75,000 and Over']


x = np.arange(len(labels))  # the label locations
width = 0.40  # the width of the bars


fig, ax = plt.subplots()
rects1 = ax.bar(x - .3, no_hs, .2, label='No High School Diploma')
rects2 = ax.bar(x - .1, hs, .2, label='High School or Equivalent')
rects3 = ax.bar(x + .1, some_col, .2, label='Some College')
rects4 = ax.bar(x + .3, bach, .2, label="Bachelor's Degree or Higher")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Population')
ax.set_title('Population of CA by Education and Income in 2008 Ages 18 to 64')
ax.set_xticks(x)
ax.set_xticklabels(labels)
#ax.set_ylimit([0,2500000]

ax.legend(loc='best', fontsize='xx-small')
plt.xticks(rotation=45)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', rotation = '90',fontsize='xx-small')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

#axes.set_ylim([0,1800000])
fig.tight_layout()
fig.savefig("output_education_attainment/education_and_income_2008_18to64")
plt.show()
```

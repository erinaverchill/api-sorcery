```python
%matplotlib notebook
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# filepaths
edattain_income = "../Resources/ca-educational-attainment-personal-income-2008-2014.csv"
```

```python
#dataframes
edattain_income_df = pd.read_csv(edattain_income, encoding="utf-8")

edattain_income_df
```

```python
#clean out age

#no age filter
#age = edattain_income_df

#over 17 data
#age = edattain_income_df[edattain_income_df.Age != '00 to 17']

# 18 to 64
age = edattain_income_df.loc[edattain_income_df['Age'] == '18 to 64']

# 65 and over
#age = edattain_income_df.loc[edattain_income_df['Age'] == '65 to 80+']

#check to see what ages are selected
age['Personal Income'].value_counts()

#set value for what year to get data from
#year = '01/01/2008 12:00:00 AM'
```

```python
#create list for the years to be selected from
years = ['01/01/2012 12:00:00 AM',
         '01/01/2013 12:00:00 AM',
         '01/01/2009 12:00:00 AM',
         '01/01/2010 12:00:00 AM',
         '01/01/2008 12:00:00 AM',
         '01/01/2014 12:00:00 AM',
         '01/01/2011 12:00:00 AM']
```

```python
#create list of polulation with bachelors degree with no income for each year.

#Create empty list to store value
tot_pop = []
no_inc = []

#loop through the data and create list of polulation with bachelors degree with no income for each year.
for x in years:
    temp_year_df = age.loc[(age['Personal Income'] == 'No Income') &
                 (age['Year'] == x) &
                 (age['Educational Attainment'] == "Bachelor's degree or higher")]
    
    tot_pop_df = age.loc[(age['Personal Income'] == 'No Income') &
                 (age['Year'] == x)]
    
    temp_tot = []
    for index, row in tot_pop_df.iterrows():
        temp_tot.append(row['Population Count'])
    tot_pop.append(sum(temp_tot))
    temp_inc =[]
    for index, row in temp_year_df.iterrows():
        temp_inc.append(row['Population Count'])
    no_inc.append(sum(temp_inc))
print(no_inc)
print(tot_pop)

```

```python
labels = [2008,2009,2010,2011,2012,2013,2014]
x_amount = len(no_inc)
x_axis = np.arange(x_amount)


x_axis = np.array(no_inc, dtype = np.float)
b = np.array(tot_pop, dtype=np.float)
percent = (x_axis/b)*100
print(percent)

plt.plot(labels, percent, label='No Income', linestyle = (0, (1, 1)))
plt.ylim(0, 100)
```

```python
import pandas as pd
import numpy as np
```

```python
epi = pd.read_csv('education_personal_income.csv')
rec = pd.read_csv('regional_economic_analysis.csv')
```

```python
epi.head()
```

```python
rec.head()
```

```python
# renaming columns to join datasets on
epi = epi.rename(columns={'Educational Attainment':'Education Level'})
rec = rec.rename(columns={'Entry Level Education':'Education Level'})
epi
```

```python
# Re-ordering rec dataframe columns
rec = rec[["Education Level","Area Type","Area Name",
           "Industry Cluster","SOC Code","SOC Title","Median Hourly Wage","Median Annual Wage","Period",
          "New Job Openings","Replacement Job Openings","Total Projected Job openings","Total Projected Job openings",
          "Online Job Ads (HWOL)"]]
rec.head()
```

```python
rec.to_csv("original_rec.csv")
```

```python
# Getting all unique values for 'Area Type' column in rec
rec_type_val = rec[["Area Type"]].values
unique_rec_type =  np.unique(rec_type_val)
print(unique_rec_type)
```

```python
# Getting all unique values for 'Area Name' column in rec
rec_name_val = rec[["Area Name"]].values
unique_rec_name =  np.unique(rec_name_val)
print(unique_rec_name)
```

```python
# Getting all unique values for 'Industry Cluster' column in rec
rec_ind_val = rec[["Industry Cluster"]].values
unique_rec_ind =  np.unique(rec_ind_val)
print(unique_rec_ind)
```

```python
# Ordering columns to merge on "Education Level" later
epi = epi[["Education Level", "Year", "Age", "Gender", "Personal Income", "Population Count"]]
```

```python
# Cleaning up 'Year' Column
epi['Year'] = pd.to_datetime(epi['Year'], errors='coerce')
# Gathering data from 2012-2014, since that's the data that corresponds with the rec dataframe
start_year = '2012-01-01'
end_year = '2014-01-01'
year_range = (epi['Year'] >= start_year) & (epi['Year'] <= end_year)
epi = epi.loc[year_range]
epi
```

```python
# Cleaning up 'Personal Income' column
epi_income_val = epi[["Personal Income"]].values
unique_epi_income =  np.unique(epi_income_val)
print(unique_epi_income)
```

```python
# cleaning up personal income values
epi = epi.replace(to_replace ="$5,000 to $9,999", value ="5,000 - 9,999") 
epi = epi.replace(to_replace="$10,000 to $14,999", value="10,000 - 14,000")
epi = epi.replace(to_replace="$15,000 to $24,999", value="15,000 - 24,999")
epi = epi.replace(to_replace="$25,000 to $34,999", value="25,000 - 34,999")
epi = epi.replace(to_replace="$35,000 to $49,999", value="35,000 - 49,999")
epi = epi.replace(to_replace="$50,000 to $74,999", value="50,000 - 74,999")
epi = epi.replace(to_replace="$75,000 and over", value="75,000 and over")
epi
```

```python
# First merged dataframe!! It's a bit massive though, so I'm going to make it more dense.
big_df = pd.merge(rec,epi,how="outer")
```

```python
# Getting epi column values
epi_column_values = epi["Education Level"].values.ravel()
epi_unique_values =  pd.unique(epi_column_values)
epi_unique_values
```

```python
# Getting rec column values
column_values = rec["Education Level"].values.ravel()
unique_values =  pd.unique(column_values)
unique_values
```

```python
# Making uniform column values, pt. 1
rec = rec.replace(to_replace="Associate's degree", value="Some college, less than 4-yr degree")
rec = rec.replace(to_replace="Some college, no degree", value="Some college, less than 4-yr degree")
rec = rec.replace(to_replace="Bachelor's degree", value="Bachelor's degree or higher")
rec = rec.replace(to_replace="Doctoral or professional degree", value="Bachelor's degree or higher")
rec = rec.replace(to_replace="Postsecondary non-degree award", value="Bachelor's degree or higher")
rec = rec.replace(to_replace="Master's degree", value="Bachelor's degree or higher")
rec.head()
```

```python
# Making uniform column values, pt. 2
epi = epi.replace(to_replace="Children under 15", value="Less than high school")
epi = epi.replace(to_replace="No high school diploma", value="Less than high school")
epi = epi.replace(to_replace="High school or equivalent", value="High school diploma or equivalent")
epi
```

```python
# Let's merge!
df = pd.merge(rec,epi,on="Education Level")
df
```

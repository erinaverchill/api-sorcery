```python
#clean out age

#no age filter
#over_17 = edattain_income_df

#over 17 data
#over_17 = edattain_income_df[edattain_income_df.Age != '00 to 17']

# create a data set for age 18 to 64 for the year 2014
over_17 = edattain_income_df.loc[(edattain_income_df['Age'] == '18 to 64') & 
            (edattain_income_df['Year'] == '01/01/2014 12:00:00 AM')]

year = "01/01/2014 12:00:00 AM"
```

```python
# create a male and female set for age 18 to 64 for the year 2014


over_17M = over_17.loc[over_17['Gender'] == 'Male']
over_17F = over_17.loc[over_17['Gender'] == 'Female']


#over_17 is male
# over_17F is female
# calculate the relative numbers of male and female and make a weighting coefficient
F_Pop = over_17F.loc[over_17F["Gender"] == "Female"]
female = F_Pop["Population Count"].sum()
M_Pop = over_17.loc[over_17["Gender"] == "Male"]
male = M_Pop["Population Count"].sum()
coeff =female/male
print(male, female)
print (coeff)
```

```python
# weight the population counts for the majority of men
over_17M["Population Count"] = round(coeff*over_17M["Population Count"], 0)
```

```python
# check the relative numbers of male and female are equa
F_Pop = over_17F.loc[over_17F["Gender"] == "Female"]
female = F_Pop["Population Count"].sum()
M_Pop = over_17M.loc[over_17M["Gender"] == "Male"]
male = M_Pop["Population Count"].sum()
print(female, male)
```
